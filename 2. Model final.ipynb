```python
import pandas as pd
import numpy as np
import gdown
import shutil
import re
import pickle
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Import file dataset
url = "https://drive.google.com/file/d/1y9qBtV71aY6OtEnhfQAxMTWXBBmdv63i/view?usp=drive_link"
output = "dataset_final.csv"
gdown.download(url, output, quiet=False, fuzzy=True)

dataset_final = pd.read_csv("dataset_final.csv")
```

```python
count = CountVectorizer()
count_matrix = count.fit_transform(dataset_final['deskripsi_keterampilan_final'])
cosine_sim = cosine_similarity(count_matrix)
print(cosine_sim)
```

```python
# Simpan model CountVectorizer ke file pickle
with open('count_vectorizer.pkl', 'wb') as f:
    pickle.dump(count, f)
```
